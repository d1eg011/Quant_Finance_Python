Recall the four principles of Object Oriented Programming:

**Abstraction:** hide non-essential information to the user.

**Encapsulation:** bundling variables and functions into a single class.

**Polymorphism:** a single object can take multiple forms depending on the problem (e.g. method overloading: same name, different parameters, different behavior; e.g. optional arguments in a function).

**Inheritance:** a sub-class uses code from a parent class.
